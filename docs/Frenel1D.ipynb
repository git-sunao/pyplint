```python
import numpy as np
import matplotlib.pyplot as plt
from pyplint.pyplint1D import Thimble
```

### Frenel integral

$$
\int_{-\infty}^{\infty} {\rm d}x e^{ix^2} = (1+i)\sqrt{\pi/2}
$$

```python
def expFrenel(x):
    return 1j*x**2
```

```python
thimble = Thimble(-2, 2)
thimble.set_phi(expFrenel)
```

```python
thimble.find()
```

```python
fig, ax = thimble.plot()
```

```python
# PL evaluation
thimble.romberg()
```

```python
# exact value
(1+1j)*np.sqrt(np.pi/2)
```
