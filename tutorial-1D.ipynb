## Picard Lefschetz integral
### 1. define the funciton on exp


```python
def phi_lens(x):
    nu = 10.0
    mu = 0.0
    o = (x-mu)**2/2 + 2/(1+x**2)
    o*= 1j*nu
    return o
```

### 2. Instantiate the thimble class

```python
thimble = pyplint.Thimble(-5, 5, delta=0.1, thre=-30, tau=0.05)
```

### 3. Set function and run to get the thimble

```python
thimble.set_phi(phi_lens)
thimble.run()
```

### 4. plot thimble

```python
# Set plot range by hand
thimble.plot((-2,2))
```
